```python
%matplotlib inline 

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_style('white')
df = pd.read_csv('auc_plot.tsv',sep='\t')
```

```python
df.head()
```

```python
p = sns.FacetGrid(data=df, col = 'base',row = 'enzyme',sharex=False)
p.map(sns.swarmplot, 'classes','AUC','type',
      palette=sns.palettes.color_palette('Set2',2),
      hue_order=np.unique(df['type']))
p.add_legend()
```
